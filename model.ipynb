# лІ”мЈјнҳ• ліҖмҲҳ мқёмҪ”л”©

```python
categorical_columns = df.select_dtypes(include=['object', 'bool']).columns

label_encoder = LabelEncoder()

# к°Ғ лІ”мЈјнҳ• ліҖмҲҳм—җ лҢҖн•ҙ л°ҳліөл¬ёмқ„ нҶөн•ң мқёмҪ”л”© мҲҳн–ү
for column in categorical_columns:
    df[column + '_encoded'] = label_encoder.fit_transform(df[column])

df
```

```python
# лІ”мЈјнҳ• ліҖмҲҳ м№јлҹј мӮӯм ң
cols_to_drop = ['property_type', 'room_type', 'bed_type', 'cancellation_policy', 'cleaning_fee', 'city']
df = df.drop(columns = cols_to_drop)
```

```python
df
```

```python
df.info()
```

# нӣҲл ЁлҚ°мқҙн„° н…ҢмҠӨнҠёлҚ°мқҙн„° л¶„н• н•ҳкё°

```python
X = df.drop('log_price', axis=1)
y = df['log_price']

# лҚ°мқҙн„°лҘј 7:3 л№„мңЁлЎң л¶„н• 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X.info()
```

# лӘЁлҚё кө¬м¶•

```python
model = Sequential()
model.add(Dense(64, activation = 'relu', input_dim = 26))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(16, activation = 'relu'))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(1, activation = 'linear'))

model.compile(optimizer = 'adam', loss = 'mean_squared_error')

# early stopping м Ғмҡ©
from keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='loss', patience=3, restore_best_weights=True)
```

# лӘЁлҚё н•ҷмҠө

```python
model.fit(X_train, y_train, epochs=50, batch_size=20, callbacks = [early_stopping])
```

# кө¬м¶•н•ң лӘЁлҚёлЎң мҳҲмёЎ м§„н–ү

```python
y_pred = model.predict(X_test)
y_pred
```

# м„ұлҠҘнҸүк°Җ

```python
mse = mean_squared_error(y_test, y_pred)
print("mse : %.4f" % (mse))
```

```python
r2 = r2_score(y_test, y_pred)
print("R^2 мҠӨмҪ”м–ҙ : %.4f" % (r2))
```

```python
corr = np.corrcoef(y_test, y_pred.flatten())[0, 1]
print("мҳҲмёЎк°’кіј мӢӨм ңк°’мқҳ мғҒкҙҖкҙҖкі„ : %.4f" % (corr))
```
